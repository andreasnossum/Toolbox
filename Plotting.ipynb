# Libraries and plotting

## Libraries

```python
# Numpy

import numpy as np

# Matplotlib

import matplotlib.pyplot as plt
from IPython.display import Image

# Scipy

from scipy import integrate
from scipy import optimize

# Astropy

import astropy.units as u
from astropy.constants import m_e, m_p, e, h, c, k_B, alpha, sigma_T
from astropy.coordinates import Distance

# Warnings

import warnings
warnings.filterwarnings("ignore", category=RuntimeWarning)
```

## Plotting

```python
x_array = np.linspace(0, 10, 100)
set_a = np.array([np.sqrt(x) for x in x_array])
set_b = np.array([np.exp(-x) for x in x_array])
set_c = np.array([0.1*x for x in x_array])
set_d = np.array([np.exp(-x) * np.cos(2*np.pi*x) for x in x_array])
```

### Default

```python
plt.figure()

#plt.style.use("ggplot")
#plt.style.use("bmh")
plt.style.use("seaborn")
#plt.style.use("fivethirtyeight")

#plt.rcParams["font.size"] = 10
#plt.rcParams["grid.color"] = "black"
plt.rcParams["figure.figsize"] = [10,6]

#plt.loglog() for logarithmix scales
plt.plot(x_array, set_a, color="r", linestyle="--", label=r"$\sqrt{x}$")
plt.plot(x_array, set_b, color="b", linestyle="-", label=r"$e^x$")
plt.plot(x_array, set_c, color="g", linestyle="-.", label=r"$0.1 x$")
plt.plot(x_array, set_d, color="m", linestyle=":", label=r"$e^{-x}\cos(2\pi x)$")

plt.annotate('Arrow', xy=(6.2, 2.3), xytext=(7.5, 1.5),
             arrowprops=dict(facecolor='black', shrink=10))

plt.title("Title")
plt.xlabel("x-axis")
plt.ylabel("y-axis", rotation=0)

#plt.xlim(-1, 11)
#plt.ylim(-1, 4)

#plt.grid()
plt.legend()

plt.show()
```

### Multiple figures

```python
fig, axs = plt.subplots(2, 2)

axs[0, 0].plot(x_array, set_a, "r")
axs[0, 0].set_title('Axis [0, 0]')

axs[0, 1].plot(x_array, set_b, "b")
axs[0, 1].set_title('Axis [0, 1]')

axs[1, 0].plot(x_array, set_c, "g")
axs[1, 0].set_title('Axis [1, 0]')

axs[1, 1].plot(x_array, set_d, "m")
axs[1, 1].set_title('Axis [1, 1]')

for ax in axs.flat:
    ax.set(xlabel='x-label', ylabel='y-label')

# Hide x labels and tick labels for top plots and y ticks for right plots.
for ax in axs.flat:
    ax.label_outer()
```

```python
fig, axs = plt.subplots(4, sharex=True, sharey=True)
fig.suptitle('Sharing both axes')
axs[0].plot(x_array, set_a, "r")
axs[1].plot(x_array, set_b, "b")
axs[2].plot(x_array, set_c, "g")
axs[3].plot(x_array, set_d, "m")
```
